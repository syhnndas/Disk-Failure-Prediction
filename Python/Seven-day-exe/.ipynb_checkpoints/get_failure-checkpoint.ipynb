```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
# 读取故障样本
df = pd.read_parquet("ST12000NM0008_all.parquet")
# 2. 日期格式转换
df['date'] = pd.to_datetime(df['date'])
# 4. 按 serial_number + date 排序
df = df.sort_values(by=['serial_number', 'date'])
# 5. 按硬盘分组处理
result_list = []
for serial, group in tqdm(df.groupby('serial_number')):
    group = group.sort_values('date').reset_index(drop=True)
    # 找到 failure = 1 的位置
    failure_indices = group.index[group['failure'] == 1].tolist()
    for idx in failure_indices:
        # 向前取6天 + 当前1天 = 7天
        start_idx = max(0, idx - 6)
        window = group.loc[start_idx:idx]
        # 只保留刚好7天的数据（可选）
        if len(window) == 7:
            result_list.append(window)
# 6. 合并结果
if result_list:
    result_df = pd.concat(result_list, ignore_index=True)
else:
    result_df = pd.DataFrame()
# 7. 保存结果
result_df.to_csv("failure_windows.csv", index=False)
print("处理完成，共提取窗口数量：", len(result_list))
```

```python
## 读取正常样本
import pandas as pd
import numpy as np
from tqdm import tqdm
# ===== 配置 =====
INPUT_FILE = "ST12000NM0008_all.parquet"
OUTPUT_FILE = "normal_windows.csv"
N_SAMPLES = 2322
# ===== 读取数据 =====
df = pd.read_parquet(INPUT_FILE)
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by=['serial_number', 'date'])
print("开始处理...")
normal_windows = []
count = 0
# ===== 单进程处理 =====
for serial, group in tqdm(df.groupby('serial_number'), desc="处理硬盘"):
    group = group.sort_values('date').reset_index(drop=True)
    candidate_indices = group.index[group['failure'] == 0].tolist()
    valid_windows = []
    for idx in candidate_indices:
        if idx < 6:
            continue
        window = group.loc[idx-6:idx]
        if window['failure'].sum() > 0:
            continue
        valid_windows.append(window)
    if len(valid_windows) > 0:
        # ⭐ 方式1：取第一个
        selected_window = valid_windows[0]
        # ⭐ 方式2（推荐）：随机选一个（更均匀）
        # selected_window = valid_windows[np.random.randint(len(valid_windows))]
        normal_windows.append(selected_window)
        count += 1
    # 满足数量就停止
    if count >= N_SAMPLES:
        break
# ===== 检查 =====
if len(normal_windows) == 0:
    print("⚠️ 没有找到正常窗口")
    exit()
# ===== 合并 =====
normal_df = pd.concat(normal_windows, ignore_index=True)
# ===== 打 window_id =====
normal_df['window_id'] = normal_df.index // 7
# ===== 保存 =====
normal_df.to_csv(OUTPUT_FILE, index=False)
print(f"✅ 完成！共采集 {count} 个正常窗口（每个硬盘最多1个）")
```
